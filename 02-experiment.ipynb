**Experiment #2**

**The question to explore is: how will changes in class balance affect AUCs when the sample size remains the same? To investigate this, we will vary `class_weight` from 40% to 0.05%, with `n_objects` set at 1E+5 and 5E+7.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'Charter'
rcParams['font.size'] = 12
rcParams['legend.fontsize'] = 12
rcParams['axes.labelpad'] = 12
rcParams['axes.labelsize'] = 12
rcParams['axes.titlesize'] = 12
rcParams['ytick.labelsize'] = 11
rcParams['xtick.labelsize'] = 11
```

```python
def get_results(n_objects):
    """Get maximum, minimum and mean AUCs
    for the specific `class_weight` and `n_objects`.
    """
    out = {
        'ROC': {'max': [], 'min': [], 'mean': []},
        'PR': {'max': [], 'min': [], 'mean': []},
    }

    for class_weight in class_weights:
        print('{:.4f} #'.format(class_weight), end='')
        roc_scores, pr_scores = [], []
        for seed in seeds:
            print('-', end='')
            roc, pr = get_metrics(
                get_sample(n_objects, class_weight, seed)
            )
            roc_scores.append(roc)
            pr_scores.append(pr)
        #
        out['ROC']['max'].append(max(roc_scores))
        out['ROC']['min'].append(min(roc_scores))
        out['ROC']['mean'].append(np.mean(roc_scores))
        #
        out['PR']['max'].append(max(pr_scores))
        out['PR']['min'].append(min(pr_scores))
        out['PR']['mean'].append(np.mean(pr_scores))
        #
        print()
        #
    return out
```

```python
def plot(out):
    figure, axis = plt.subplots(1, 2, figsize=(14, 7))
    #
    for idx, metric in enumerate(('ROC', 'PR')):
        axis[idx].fill_between(range(len(class_weights)), y1=out[metric]['max'], y2=out[metric]['min'], alpha=0.2, color='silver')
        axis[idx].plot(range(len(class_weights)), out[metric]['mean'], 'ko--', lw=1.3, mfc='white')
        axis[idx].grid(True, alpha=0.3, aa=True, ls=':', lw=1.1)
        axis[idx].set_ylabel('AUC {}'.format(metric))
        axis[idx].set_xlabel('Class weight')
        axis[idx].set_xticks(range(len(class_weights)))
        _ = axis[idx].set_xticklabels(['{:.2f}%'.format(cw*100) for cw in class_weights], rotation=45)
    plt.tight_layout(pad=2)
```

# `number_of_objects` = 50,000,000

```python
plot(get_results(n_objects=5e+7))
plt.savefig('figures/fig8.png', dpi=300)
```
